```python
import random
import pandas as pd
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
```

```python
#Dados carro 5
df = pd.read_json('telemetry.json', lines=True)
car5 = json_normalize( df[df['Car'] ==5 ]['telemetry'] )
car5.head()
```

```python
#RPM do carro 5
rpm = car5['RPM']

#Marcha do carro 5
gear = car5['Gear']

fig, ax1 = plt.subplots()
color = 'tab:blue'

#Grafico RPM
ax1.set_xlabel('Tempo (segundos)')
ax1.set_ylabel('RPM', color=color)
ax1.plot(rpm, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

#Grafico Gear
color = 'tab:orange'
ax2.set_ylabel('Gear', color=color)
ax2.plot(gear, color=color, linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

#Grafico final
fig.tight_layout()
plt.show()
```
