```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
```

```python
#Leitura do arquivo
df = pd.read_json('telemetry.json', lines=True)
```

```python
#Dados carro 4
car4 =  json_normalize( df[df['Car']==4]['telemetry'] )
car4.head()
```

```python
#Dados carro 5
car5 = json_normalize( df[df['Car']==5]['telemetry'] )
car5.head()
```

```python
# Grafico comparando velocidade do carro 4 e 5 em Km/h
plt.plot(car4['Speed'], color='blue')
plt.plot(car5['Speed'], color='orange')
plt.xlabel('Tempo (segundos)')
plt.ylabel('Velocidade (Km/h)')
```
