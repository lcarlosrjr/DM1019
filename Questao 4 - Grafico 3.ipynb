```python
import random
import pandas as pd
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
```

```python
#Dados carro 5
df = pd.read_json('telemetry.json', lines=True)
car5 = json_normalize( df[df['Car'] ==5 ]['telemetry'] )
car5.head()
```

```python
#Tempo da volta na pista Carro 5
distance = car5['Distance']

#Combustivel carro 5
fuel = car5['Fuel']

#Gráficos do tempo de volta em relação ao ombustível
fig, ax1 = plt.subplots()
color = 'tab:red'

#Distancia
ax1.set_xlabel('Tempo (segundos)')
ax1.set_ylabel('Distancia (metros)', color=color)
ax1.plot(distance, color=color, linewidth=2)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

#Combustivel
color = 'tab:green'
ax2.set_ylabel('Combustivel', color=color)
ax2.plot(fuel, color=color, linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

#Grafico final
fig.tight_layout()
plt.show()
```

```python
# Justificativa:

# Com o gráfico pode-se perceber que conforme a distância percorrida pelo carro vai aumentando, a quantidade de combustivel
# também diminui.
# O carro no inicio da corrida começa obvimente com o tanque cheio e conforme vai andando, a quantidade de combustível vai
# diminuindo. Nota-se também que o carro nao parou para reabaster.
```
